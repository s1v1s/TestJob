```python
# Отделяем лист "Задание1" и возвращаем одноуровневые заголовки
task1_df = all_sheets_df_dict[task_list[0]]
task1_df.columns = task1_df.columns.get_level_values(0)
# Отделяем отчёты в словарь отчётов
reports_df_dict = {
    sheet_name: all_sheets_df_dict[sheet_name] for sheet_name in report_list
}
# Отделяем дополнительную информацию в словарь и возвращаем одноуровневые заголовки
info_df_dict = {}
for sheet_name in info_list:
    info_df_dict[sheet_name] = all_sheets_df_dict[sheet_name].reset_index(drop=True)
    info_df_dict[sheet_name].columns = info_df_dict[
        sheet_name
    ].columns.get_level_values(0)
```

Цель следующей ячейки: упростить дальнейшую работу с заголовками и определением ИП

```python
def process_report_df(df, new_header):
    """Изменить ячейку с информацией об отчёте, вернуть оригинальное значение и обновить заголовок."""
    if "Отчёт по данным поставщика" in df.columns.values[0][0]:
        info_header = df.columns.values[0][0]
        df.rename(columns={info_header: new_header}, inplace=True)
        return info_header
    return None


# Новое значение для ячейки с информацией об отчёте
new_short_info_header = "Товары"

# Добавляем информацию об отчёте как атрибут для каждого датафрейма
for sheet_name, df in reports_df_dict.items():
    info_header = process_report_df(df, new_short_info_header)
    setattr(df, "info_header", info_header)
```

```python
for sheet_name in report_list:
    print(f"{sheet_name}: {reports_df_dict[sheet_name].columns.to_list()}")
```

## 1.1 Определить к какому ИП относится номенклатура.

Примечание: если номенклатура не встречается ни в одном из отчетов, в поле «ИП» поставить прочерк.

```python
for report in report_list:
    print(reports_df_dict[report].info_header)
```

```python
def find_ip_name(info):
    """Извлечь название ИП из информации отчета."""
    match = re.search(r"Отчёт по данным поставщика «(.*?) ИП»", info)
    return match.group(1) if match else "-"


def find_source(nom):
    """Найти ИП по наименованию номенклатуры."""
    for df in reports_df_dict.values():
        if nom in df[(new_short_info_header, "Номенклатура")].values:
            return find_ip_name(df.info_header)
    return "-"


# Применяем функцию для извлечения ИП в столбец "ИП"
task1_df["ИП"] = task1_df[("Номенклатура")].apply(find_source)
task1_df.head()
```

## 1.2 Посчитать количество заказов по каждой номенклатуре.

Дополнительные сведения для выполнения задания:
- данные по заказам представлены в Отчетах о Продажах по трем ИП в столбцах «Заказано, шт.»;
- если ни в одном отчете о продажах нет информации по номенклатуре, то количество заказов по такой номенклатуре равняется нулю;
- так как отчет может содержать несколько строк по одной номенклатуре, важно считать сумму заказов по всем строкам, в которых указана номенклатура.

```python
def find_number_orders(nom):
    """Суммировать кол-во заказов по заданной номенклатуре."""
    total_orders = 0
    for df in reports_df_dict.values():
        # Проверяем наличие столбца ("Заказано", "шт") и номенклатуры в отчёте
        if ("Заказано", "шт") in df.columns:
            # Суммируем кол-во заказов
            total_orders += df[df[("Товары", "Номенклатура")] == nom][
                ("Заказано", "шт")
            ].sum()
    return total_orders


# Применяем функцию к каждой номенклатуре и записываем результат в новый столбец
task1_df["Заказы, шт."] = task1_df["Номенклатура"].apply(find_number_orders)
task1_df.head()
```

## 1.3 Посчитать выручку по каждой номенклатуре.

Дополнительные сведения для выполнения задания:
- исходные данные: Отчеты о Продажах по трем ИП, столбцы «Заказано, себестоимость»;
- себестоимость в отчетах о продажах представляет собой выручку за вычетом комиссии WB (17% от выручки), т.е. себестоимость составляет 83% от выручки.

```python
# Комиссия WB
commission_rate_WB = 0.17
```

```python
def find_revenue(nom, commission):
    """Вычислить общую выручку для заданной номенклатуры с учетом комиссии."""
    total_revenue = 0
    for df in reports_df_dict.values():
        # Проверяем наличие необходимых столбцов в отчёте
        if (
            "Заказано",
            "себестоимость",
        ) in df.columns and (
            "Товары",
            "Номенклатура",
        ) in df.columns:
            # Фильтруем данные по номенклатуре и рассчитываем выручку
            revenue = df.loc[
                df[("Товары", "Номенклатура")] == nom, ("Заказано", "себестоимость")
            ]
            total_revenue += round(revenue.sum() / (1 - commission), 2)
    return total_revenue


# Применяем функцию к каждой номенклатуре и записываем результат в новый столбец
task1_df["Выручка, руб."] = task1_df["Номенклатура"].apply(
    find_revenue, commission=commission_rate_WB
)
task1_df.head()
```

## 1.4 Посчитать прибыль по каждой номенклатуре.

Дополнительные сведения для выполнения задания:
- исходные данные: выручка, полученная в предыдущем пункте, таблица себестоимости и справочник;
- прибыль=выручка-затраты;
- затраты включают фиксированные затраты (на шт., не зависят от цены товара), налог (ИП Иванов – 1%, ИП Петров – 3%, ИП Сидоров – 5%) и комиссию (17%); налог и комиссия вычисляются как процент от выручки;


Примечание: если себестоимость не указана для какой-то из категорий, то задаем ее равной минимальной себестоимости из таблицы с себестоимостями.

```python
print(f"{task_list[0]}, Список 'ИП': {task1_df["ИП"].unique()}")
print(f"{task_list[0]}: {task1_df.columns.to_list()}")
for sheet_name in info_list:
    print(f"{sheet_name}: {info_df_dict[sheet_name].columns.to_list()}")
```

```python
# Налог для каждого ИП
tax_rates = {
    "Иванов Иван": 0.01,
    "Петров Александр": 0.03,
    "Сидоров Анатолий": 0.05,
}
# Комиссия WB задана ранее в commission_rate_WB
```

```python
# Объединяем таблицы "Задание1" "Справочник" и "Себестоимость и Цена до скидки"
merged_df = task1_df.merge(
    info_df_dict["Справочник"], on="Номенклатура", how="left"
).merge(info_df_dict["Себестоимость и Цена до скидки"], on="Категория", how="left")

# Заполняем отсутствующие фиксированные затраты минимальными значениями
min_fixed_cost = info_df_dict["Себестоимость и Цена до скидки"][
    "Фиксированные затраты, руб./шт."
].min()
merged_df["Фиксированные затраты, руб./шт."] = merged_df[
    "Фиксированные затраты, руб./шт."
].fillna(min_fixed_cost)

# Рассчитываем затраты и прибыль
merged_df["Фиксированные затраты, руб."] = (
    merged_df["Фиксированные затраты, руб./шт."] * merged_df["Заказы, шт."]
)
merged_df["Налог, руб."] = merged_df["Выручка, руб."] * merged_df["ИП"].map(tax_rates)
merged_df["Комиссия, руб."] = merged_df["Выручка, руб."] * commission_rate_WB

# Рассчитываем прибыль
merged_df["Прибыль, руб."] = merged_df["Выручка, руб."] - (
    merged_df[["Фиксированные затраты, руб.", "Налог, руб.", "Комиссия, руб."]].sum(
        axis=1
    )
)

# Обновляем исходный DataFrame task1_df, добавляя новый столбец "Прибыль, руб."
task1_df["Прибыль, руб."] = merged_df["Прибыль, руб."]
task1_df.head()
```

## 1.5 Посчитать рентабельность продаж по каждой номенклатуре.

(Рентабельность – соотношение прибыли к выручке; в процентах).

```python
task1_df["Рентабельность, %"] = (
    task1_df["Прибыль, руб."] / task1_df["Выручка, руб."]
) * 100
task1_df["Рентабельность, %"] = task1_df["Рентабельность, %"].round(2)
task1_df.head()
```

## 2.1 Подтянуть Цену до скидки и Необходимую скидку для участия в акции.

```python
# Объединение для подтягивания "Категория" в task2_df
task2_df = task2_df.merge(info_df_dict["Справочник"], on="Номенклатура", how="left")
task2_df.head()
```

```python
# Подтягивание "Базовая цена" из таблицы "Себестоимость и Цена до скидки"
task2_df = task2_df.merge(
    info_df_dict["Себестоимость и Цена до скидки"][["Категория", "Базовая цена"]],
    on="Категория",
    how="left",
)
# Заполнение существующего "Цена до скидки" данными из "Базовая цена"
task2_df["Цена до скидки"] = task2_df["Цена до скидки"].fillna(task2_df["Базовая цена"])
task2_df.drop(columns=["Базовая цена"], inplace=True)

task2_df.head()
```

```python
# Подтягивание "Необходимая скидка для участия в акции" из таблицы "Акция"
task2_df = task2_df.merge(
    info_df_dict["Акция"][
        ["Номенклатура", "Необходимая скидка для участия в акции, %"]
    ],
    on="Номенклатура",
    how="left",
)
# Заполнение существующего  "Необходимая скидка для участия в акции" данными из "Необходимая скидка для участия в акции, %"
task2_df["Необходимая скидка для участия в акции"] = task2_df[
    "Необходимая скидка для участия в акции"
].fillna(task2_df["Необходимая скидка для участия в акции, %"])
task2_df.drop(columns=["Необходимая скидка для участия в акции, %"], inplace=True)
task2_df.head()
```

## 2.2 Рассчитать цену после скидки.

```python
task2_df["Цена после скидки"] = task2_df["Цена до скидки"] * (
    1 - task2_df["Необходимая скидка для участия в акции"] / 100
)
task2_df.head()
```

## 2.3 Подтянуть Фиксированные затраты.

```python
# Подтягивание "Фиксированные затраты, руб./шт." из таблицы "Себестоимость и Цена до скидки"
task2_df = task2_df.merge(
    info_df_dict["Себестоимость и Цена до скидки"][
        ["Категория", "Фиксированные затраты, руб./шт."]
    ],
    on="Категория",
    how="left",
)
# Заполнение существующего "Фиксированные затрат, руб./шт." данными из "Фиксированные затраты, руб./шт."
task2_df["Фиксированные затрат, руб./шт."] = task2_df[
    "Фиксированные затраты, руб./шт."
].fillna(task2_df["Фиксированные затраты, руб./шт."])
task2_df.drop(columns=["Фиксированные затраты, руб./шт."], inplace=True)

task2_df.head()
```

## 2.4 Налоговую ставку задать как случайное целое число от 1 до 6%.

```python
task2_df["Налог, %"] = np.random.randint(1, 7, size=len(task2_df))

task2_df.head()
```

## 2.5 Указать комиссию в зависимости от модели продаж (FBS – 10%, FBO – 15%).

```python
# Установка комиссии в зависимости от модели продаж
task2_df["Комиссия,% "] = task2_df["Модель продаж"].apply(
    lambda x: 10 if x == "FBS" else (15 if x == "FBO" else None)
)
task2_df.head()
```

## 2.6 Посчитать рентабельность.

- Рентабельность = (Цена после скидки-Затраты на ед.)/Цена после скидки
- Затраты на ед.=Фиксированные затраты+(Налог + Комиссия)*Цена после скидки.

```python
# Рассчитываем затраты на единицу
task2_df["Затраты на ед."] = task2_df["Фиксированные затрат, руб./шт."] + (
    (task2_df["Налог, %"] + task2_df["Комиссия,% "])
    / 100
    * task2_df["Цена после скидки"]
).round(2)

# Рассчитываем рентабельность
task2_df["Рентабельность, %"] = (
    (task2_df["Цена после скидки"] - task2_df["Затраты на ед."])
    / task2_df["Цена после скидки"]
    * 100
).round(2)

task2_df.head()
```

## 2.7 Если рентабельность меньше 10%, то в столбце Проверка выводим False, иначе True.

```python
# Условное заполнение столбца "Проверка"
task2_df["Проверка"] = task2_df["Рентабельность, %"].apply(
    lambda x: False if x < 10 else True
)
task2_df.head()
```

## 2.8 Привести к нужным типам данных:

- Цена до скидки, Необходимая цена для участия в акции, Цена после скидки – целые;

- Фиксированные затраты, Налог, Комиссия, Рентабельность – числа с плавающей запятой;

- Проверка – строка.

```python
# Заполнение пустых значений и приведение типов данных
columns_to_int = [
    "Цена до скидки",
    "Необходимая скидка для участия в акции",
    "Цена после скидки",
]
columns_to_float = [
    "Фиксированные затрат, руб./шт.",
    "Налог, %",
    "Комиссия,% ",
    "Рентабельность, %",
]

# Заполняем NaN нулями и применяем int тип данных
task2_df[columns_to_int] = task2_df[columns_to_int].fillna(0).astype(int)

# Заполняем nan нулями и применяем float тип данных
task2_df[columns_to_float] = task2_df[columns_to_float].fillna(0).astype(float)

# Приведение столбца "Проверка" к строковому типу
task2_df["Проверка"] = task2_df["Проверка"].astype(str)

task2_df.head()
```
